```python
def find_closet_match(test_str, checklist):
    scores = {}
    for i in checklist:
        cnt = 0
        if len(test_str)<=len(i):
            firststr, secondstr = test_str, i
        else:
            firststr, secondstr = i, test_str
        for j in range(len(firststr)):
            cnt += 1 if firststr[j]==secondstr[j] else 0
        scores[i] = cnt
    scores_values = np.array(list(scores.values()))
    closest_match_idx = np.argsort(scores_values, axis=0, kind='quicksort')[-1]
    closest_matches = np.array(list(scores.keys()))[closest_match_idx]
    return closest_matches, closest_match_idx
```

```python
close,_=find_closet_match(test_str= str(input('Enter Attraction name:')) ,checklist= name_list) 
print(close)
similar_list = model.docvecs.most_similar(close)
for i in similar_list:
    print(i)
```
